```python
import time
```

```python
class xo:
  def __init__(self):
    self.initialize_game()

  def initialize_game(self):
      self.current_state = [['_','_','_'],
                            ['_','_','_'],
                            ['_','_','_']]

      self.player_turn = 'X'
    #генерация поля
  def pole(self):
      for i in range(0,3):
        for j in range(0,3):
          print('{}|'.format(self.current_state[i][j]), end = ' ')
        print()
      print()
  #проверка хода на адекватность
  def is_valid(self, px, py):
    if px < 0 or px > 2 or py < 0 or py > 2:
      return False
    elif self.current_state[px][py] != '_':
      return False
    elif px != int or py != int:
      print("Введите цифры!")
    else:
      return True
    
  #проверка на конец игры
  def is_end(self):
    #проверка по главной диагонали
    if (self.current_state[0][0] != '_' and
      self.current_state[0][0] == self.current_state[1][1] and
      self.current_state[0][0] == self.current_state[2][2]):
      return self.current_state[0][0]
    #проверка по второй диагонали
    if (self.current_state[0][2] != '_' and
      self.current_state[0][2] == self.current_state[1][1] and
      self.current_state[0][2] == self.current_state[2][0]):
      return self.current_state[0][2]
    #вертикальная проверка
    for i in range(0, 3):
      if (self.current_state[0][i] != '_' and
        self.current_state[0][i] == self.current_state[1][i] and
        self.current_state[1][i] == self.current_state[2][i]):
        return self.current_state[0][i]
    #горизонтальная проверка
    for i in range(0, 3):
      if (self.current_state[i] == ['X', 'X', 'X']):
        return 'X'
      elif (self.current_state[i] == ['O', 'O', 'O']):
        return 'O'
    #проверка полноты поля
    for i in range(0, 3):
      for j in range(0, 3):
        if (self.current_state[i][j] == '_'):
          return None
    #ничья
    return '_'

  def max(self):
    #возможные значения для maxvalue:
    # -1 - проигрыш; 0 - ничья; #1 - победа 
    #для максимизации даем значение изначальное в -2 
    maxvalue = -2

    px , py = None, None

    result = self.is_end()

    if result =='X':
      return(-1,0,0)
    elif result=='O':
      return(1,0,0)
    elif result == '_':
      return(0,0,0)
    
    for i in range(0,3):
      for j in range(0,3):
        if self.current_state[i][j] =='_':
          #на пустом поле игрок за О делает ход и вызывает функцию min
          #это один из веток игры
          self.current_state[i][j] == 'O'
          (m, min_i, min_j) = self.min()

          #исправляется значение по надобности
          if m> maxvalue:
            maxvalue = m
            px = i
            py = j

          #обнуление поля
          self.current_state[i][j] = '_'
    return(maxvalue, px, py)

  def max_alpha_beta(self, alpha, beta):
    maxvalue = -2
    px, py = None, None
    result = self.is_end()

    if result == 'X':
        return (-1, 0, 0)
    elif result == 'O':
        return (1, 0, 0)
    elif result == '_':
        return (0, 0, 0)

    for i in range(0, 3):
      for j in range(0, 3):
        if self.current_state[i][j] == '_':
          self.current_state[i][j] = 'O'
          (m, min_i, in_j) = self.min_alpha_beta(alpha, beta)
          if m > maxvalue:
            maxvalue = m
            px = i
            py = j
          self.current_state[i][j] = '_'

  #всего лишь две следуюшие циклы if отличают обычный алгоритм от 
    #алгоритма с альфа-бета обрезкой
          if maxvalue >= beta:
            return (maxvalue, px, py)
          if maxvalue > alpha:
            alpha = maxvalue

    return (maxvalue, px, py)

  def min_alpha_beta(self, alpha, beta):
    minvalue = 2

    qx = None
    qy = None

    result = self.is_end()

    if result == 'X':
        return (-1, 0, 0)
    elif result == 'O':
        return (1, 0, 0)
    elif result == '_':
        return (0, 0, 0)

    for i in range(0, 3):
      for j in range(0, 3):
        if self.current_state[i][j] == '_':
          self.current_state[i][j] = 'X'
          (m, max_i, max_j) = self.max_alpha_beta(alpha, beta)
          if m < minvalue:
            minvalue = m
            qx = i
            qy = j
          self.current_state[i][j] = '_'

          if minvalue <= alpha:
            return (minvalue, qx, qy)
          if minvalue < beta:
            beta = minvalue

    return (minvalue, qx, qy)
  def min(self):

    minvalue = 2

    qx, qy = None, None

    result = self.is_end()

    if result =='X':
      return(-1,0,0)
    elif result=='O':
      return(1,0,0)
    elif result == '_':
      return(0,0,0)

    for i in range(0,3):
      for j in range(0,3):
        if self.current_state[i][j] =='_':
          self.current_state[i][j] == 'X'
          (m, max_i, max_j) = self.max()
          if m< minvalue:
            minvalue = m
            qx = i
            qy = j
          self.current_state[i][j] = '_'
    return(minvalue, qx, qy)

  def igra(self):
    while True:
      self.pole()
      self.result = self.is_end()

      #Тексты на конец игры 
      if self.result != None:
        if self.result == 'X':
          print('Победа за игроком крестика Х!')
        elif self.result == 'O':
          print('Победа за игроком нолика O!')
        elif self.result == '_':
          print('Ничья восторжествует!')

        self.initialize_game()
        return

      if self.player_turn == 'X':

        while True:
          start = time.time()
          (m, qx, qy) = self.min()
          end = time.time()

          print('Время обработки {} секунд'.format(round(end - start, 7)))
          print('Рекомендуемые ходы: X = {} Y = {}'.format(qx, qy))

          px = int(input('Введите координаты X: '))
          py = int(input('Введите координаты Y: '))

          (qx, qy) = (px, py)

          if self.is_valid(px, py):
            self.current_state[px][py] = 'X'
            self.player_turn = 'O'
            break
          else:
            print('Ход не адекватен. Введите еще раз')
      else:
        (m, px, py) = self.max()
        self.current_state[px][py] = 'O'
        self.player_turn = 'X'
```

```python
def main():
    g = xo()
    g.igra()

if __name__ == "__main__":
    main()
```
